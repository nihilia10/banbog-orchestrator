# BanBog Orchestrator - Exploration Notebook

This notebook demonstrates how to use the BanBog Orchestrator Agent to query products, reviews, and technical documentation.

```python
import os
import sys
from dotenv import load_dotenv

# Add the parent directory to sys.path to import the src package
sys.path.append(os.path.abspath('..'))

from src.orchestrator_agent import create_orchestrator

# Load environment variables
load_dotenv('../.env')

print("System initialized.")
```

```python
# Initialize the orchestrator
orchestrator = create_orchestrator()

def ask(question):
    print(f"USER: {question}")
    answer = orchestrator.invoke(question)
    print(f"AGENT: {answer}\n")
    print("-" * 50)
```

```python
# Test: Products
ask("¿Cuáles son los beneficios de la tarjeta de crédito?")
```

```python
# Test: Reviews (SQL Agent)
ask("¿Quién es el usuario que más reseñas ha realizado?")
```

```python
# Test: BRE-B (Technical documentation)
ask("¿Qué es el sistema BRE-B?")
```
